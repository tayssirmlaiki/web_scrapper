```python
import requests
from bs4 import BeautifulSoup 
import re
import pandas
```

```python
Link ="https://www.mosaiquefm.net/fr/actualite-economie-tunisie/512466/reprise-de-la-cotation-du-titre-attijari-bank.html"
#convert the web page to text
Link_text = requests.get(Link).text
print(Link_text)
```

```python
soup = BeautifulSoup(Link_text, 'lxml')
print(soup)
```

```python
#make the indentation proper
print(soup.prettify())
```

```python
#To take a look at the title of the web page
print(soup.title)
```

```python
#Only the string not the tags
title =soup.title.string
print(title)
```

```python
#First <a></a> tag
soup.a
```

```python
#all the <a> </a> tags
soup.find_all('a')
```

```python
#Fetch all the table tags
all_table = soup.find_all('table')
print(all_table)
```

```python
#all the <a> </a> tags
description = soup.body.find_all('p')
function_usage =[]

for item in description:
    item = item.text
   
    function_usage.append(item)
    
print(function_usage)
```

```python
#creat a dataFrame
data = pd.DataFrame({'function Title':title,'function description':function_usage})
data
```

```python
data.to_csv('my_file.csv')
```
